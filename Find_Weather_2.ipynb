#### __Test__
- __ALLSKY_SFC_UV_INDEX:__ All sky surface UV index.
- __WS2M:__ The average of wind speed at 2 meters above the surface of the earth.
- __PS:__ The average of surface pressure at the surface of the earth.
- __ALLSKY_SFC_UVA & ALLSKY_SFC_UVB:__ UV-A and UV-B. 
- UVA的意思就是长波紫外线 波长介于320-400纳米，可穿透云层、玻璃进入室内及车内，会射入皮肤的真皮层，造成晒红和晒伤，也是皮肤老化和出现皱纹的主因。UVB的意思是中波紫外线 波长介于290-320纳米，会被平流层的臭氧所吸收，会引起晒伤及皮肤红肿热痛，UVB致癌性最强。

```python
url_empty =  "https://power.larc.nasa.gov/api/temporal/daily/point?parameters=ALLSKY_SFC_UV_INDEX,WS2M,PS,ALLSKY_SFC_UVA,ALLSKY_SFC_UVB&community=AG&longitude={}&latitude={}&start=20130101&end=20201231&format=CSV"
```

```python
error_ls = []
start = 0  # The max number within existing files' name
end = num
```

```python
for i_ in range(start, end):
    lat = df.iloc[i_, 2]
    lon = df.iloc[i_, 3]
    url_ = url_empty.format(lon, lat)
    file_name = '{}.csv'.format(int(df.iloc[i_, 4]))
    
                                
    try:
        f = requests.get(url_)
    except Exception:
        print("Retry·············")  # Alert
        try:
            sleep(9 + np.random.normal(loc=3.0, scale=1.2))
            f = requests.get(url_)
        except Exception:
            error_ls.append(i_)
            continue
        else:
            pass
    

    with open(file_name, "wb") as file_content:
        file_content.write(f.content)
            
    with open(file_name, "r") as test:
        c = test.read()
        
    if c == "<html><body><h1>408 Request Time-out</h1>\nYour browser didn't send a complete request in time.\n</body></html>\n":
        print("Retry·············")  # Alert
        try:
            f = requests.get(url_)
        except Exception:
            error_ls.append(i_)
            continue
        else:
            with open(file_name, "wb") as file_content:
                file_content.write(f.content)
            
            
    sleep(10 + np.random.normal(loc=.0, scale=1.2))
    
```
